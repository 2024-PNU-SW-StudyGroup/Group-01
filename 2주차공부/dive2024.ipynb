```python
import pandas as pd

data = pd.read_csv('/Users/wonjds/Documents/python/data.csv')
data.head(10)
```

```python
filtered_bq1 = data[(data['CQ6'] >= 3) & (data['CQ5_1'] == 3)]['BQ1']
print(filtered_bq1.head())
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Map the values in sq1 and sq2 columns
data['Gender'] = data['SQ1'].map({1: 'Male', 2: 'Female'})
data['AgeGroup'] = data['SQ2'].map({2: '20s', 3: '30s'})

# Melt the dataframe to long format for easier plotting
melted_data = data.melt(id_vars=['Gender', 'AgeGroup'], value_vars=['BQ10_1', 'BQ10_2', 'BQ10_3', 'BQ10_4', 'BQ10_5', 'BQ10_6', 'BQ10_7'], 
                        var_name='Question', value_name='Response')
# Drop rows with NaN values in the 'Response' column
melted_data = melted_data.dropna(subset=['Response'])
# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x='Question', y='Response', hue='Gender', data=melted_data, ci=None)
plt.title('Responses by Gender and Age Group')
plt.show()
```

```python
# Filter the data for 20s and 30s age groups
filtered_data = melted_data[melted_data['AgeGroup'].isin(['20s', '30s'])]

# Plot the data
plt.figure(figsize=(12, 6))
sns.barplot(x='Question', y='Response', hue='AgeGroup', data=filtered_data, ci=None)
plt.title('Responses by Age Group (20s vs 30s)')
plt.show()
```
